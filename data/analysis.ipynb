```python
import pandas as pd
```

```python
data = pd.read_csv('./train.csv')
```

```python
total = {}
for t1,t2 in zip(data['title1_zh'],data['title2_zh']):
    try:
        l = len(t1)+len(t2)
    except:
        l = len(t1)
    try:
        total[l] += 1
    except:
        total[l] = 1
```

```python
for key in sorted(total.keys()):
    print(key,total[key])
```

```python
diff = [{},{},{}]
count = [0,0,0]
for t1,t2,l in zip(data['title1_zh'],data['title2_zh'],data['label']):
    s1 = set(t1)
    if(type(t2)==float):
        t2 = ''
    s2 = set(t2)
    t = (s1 | s2) - (s1 & s2)
    
    if(l=='unrelated'):
        l=0
    elif(l=='agreed'):
        l=1
    elif(l=='disagreed'):
        l=2
    else:
        print('error')
    count[l] +=1
    
    for n in t:
        try:
            diff[l][n] += 1
        except:
            diff[l][n] = 1
    
```
