```python
import pandas as pd
```

```python
data = pd.read_csv('./train.csv')
```

```python
total = {}
for t1,t2 in zip(data['title1_zh'],data['title2_zh']):
    try:
        l = len(t1)+len(t2)
    except:
        l = len(t1)
    try:
        total[l] += 1
    except:
        total[l] = 1
    if(l>100):
        print([t1],[t2])
```

```python
for key in sorted(total.keys()):
    print(key,total[key])
```
